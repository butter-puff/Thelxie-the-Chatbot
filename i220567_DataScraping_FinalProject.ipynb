```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import time


dfs = []


for page in range(0,4):
    for i in range(0, 40):
        # Redefine grid
        phone_listings = driver.find_element(By.XPATH, '/html/body/div[3]/div/div[3]/div/div/div[1]/div[2]')
        phones = phone_listings.find_elements(By.CLASS_NAME, 'gridItem--Yd0sa')

     
        phone_url = driver.current_url

        phones[i].click()
        time.sleep(5)

        title = driver.find_elements(By.CLASS_NAME, 'pdp-mod-product-badge-wrapper')
        brand = driver.find_elements(By.CLASS_NAME, 'pdp-product-brand')
        price = driver.find_elements(By.CLASS_NAME, 'pdp-product-price')
        rating = driver.find_elements(By.XPATH, '/html/body/div[4]/div/div[3]/div[2]/div/div[1]/div[4]/div[1]/div/div/a')

     
        titles = [t.text for t in title]
        brands = [b.text for b in brand]
        prices = [p.text for p in price]
        ratings = [r.text for r in rating]

        
        max_length = max(len(titles), len(brands), len(prices), len(ratings))

     
        titles += [None] * (max_length - len(titles))
        brands += [None] * (max_length - len(brands))
        prices += [None] * (max_length - len(prices))
        ratings += [None] * (max_length - len(ratings))

       
        driver.execute_script("window.scrollTo(0, 500);")
        time.sleep(1)
        driver.execute_script("window.scrollTo(500, 800);")
        time.sleep(1)
        driver.execute_script("window.scrollTo(800, 1000);")
        time.sleep(2)
        driver.execute_script("window.scrollTo(1000, 1300);")
        time.sleep(6)

        five = driver.find_elements(By.CLASS_NAME, 'score')
        scores = [f.text for f in five]

     
        reviews = []
        for page in range(0, 3):
            review_elements = driver.find_elements(By.CLASS_NAME, 'review-content-sl')
            reviews.extend([f"the reviews are: {r.text}\n" for r in review_elements])
            try:
                if page != 2:
                    nxt_button = driver.find_element(By.CLASS_NAME, 'ant-pagination-next')
                    driver.execute_script("arguments[0].click();", nxt_button)
                    time.sleep(5)
            except NoSuchElementException:
                print('No More Pages')
                break

            print(".............................")

       
        reviews += [None] * (max_length - len(reviews))

        # Append data to the list
        product_data = {
            'URL': [phone_url] * max_length,
            'Title': titles,
            'Brand': brands,
            'Price': prices,
            'Rating': ratings,
            'Scores': scores,
            'Reviews': reviews
        }
        dfs.append(product_data)

   
        driver.back()
    nxt_button = driver.find_element(By.XPATH, '/html/body/div[3]/div/div[3]/div/div/div[1]/div[3]/div[2]/div/ul/li[9]/a')
    driver.execute_script("arguments[0].click();", nxt_button)
    time.sleep(3)
    print("Proceeding to next page")
        
    


final_df = pd.DataFrame(dfs)


final_df.to_csv('product_data.csv', index=False)


driver.quit()
```

```python
import pandas as pd
path = r"C:\Users\Jannat\Downloads\product_data.csv"
df = pd.read_csv(path)
```

```python
import uuid
def generate_unique_id():
    return str(uuid.uuid4())
df['product_id'] = df.apply(lambda row: generate_unique_id(), axis=1)


def remove_square_brackets(value):
    return str(value).replace('[', '').replace(']', '').replace("'", '')

df = df.applymap(remove_square_brackets)

df['Brand'] = df['Brand'].str.replace('Brand:', '').str.strip()

df['Brand'] = df['Brand'].str.replace('Brand:', '').str.replace(r'More Mobiles.*','').str.strip()
df['Brand'] = df['Brand'].str.replace('Brand:', '').str.replace('AppleMore Like New Phones from Apple', 'Apple').str.strip()
df['Brand'] = df['Brand'].str.replace('Brand:', '').str.replace('AppleApple', 'Apple').str.strip()

def extract_first_word(title):
    return title.split()[0]


df.loc[df['Brand'] == 'No Brand', 'Brand'] = df.loc[df['Brand'] == 'No Brand', 'Title'].apply(extract_first_word)

df['Reviews'] = df['Reviews'].str.replace('the reviews are:', '').str.strip()


df['Price'] = df['Price'].astype(str)

df['Price'] = df['Price'].str.extract(r'Rs\. (\d{1,3}(?:,\d{3})*(?:\.\d+)?)').fillna('0')
df['Price'] = pd.to_numeric(df['Price'].str.replace(',', ''), errors='coerce')



df.to_csv('cleandata.csv', index=False)
df
```

```python
df2 = df.copy()
df2.rename(columns={'Rating': 'Num of Ratings'}, inplace=True)


df2['Answered Qs'] = df2['Num of Ratings'].copy()
df2['Num of Ratings'] = df2['Num of Ratings'].str.extract('(\d+)').fillna(0).astype(int)

df2
df2.to_csv('cleandata3.csv', index=False)
df2
```

```python
df3 = df2.copy()


df3['Answered Qs'] = df3['Answered Qs'].str.extract(r'(\d+ Answered Questions)').fillna('0 Answered Questions')


df3['Answered Qs'] = df3['Answered Qs'].str.extract(r'(\d+)').fillna(0).astype(int)
df3['Scores'].replace('', pd.NA, inplace=True)


df3['Scores'] = pd.to_numeric(df3['Scores'], errors='coerce')

df3
```

```python

df3.to_csv('finaldatatoprocess.csv', index=False)
```
